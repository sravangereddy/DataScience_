```python
dropped_cols=[]
for  i in in_master_dataset.columns : 
#     print(i)
    if in_master_dataset[i].nunique == 1 :
        in_master_dataset.drop(i,axis=1,inplace=True)
        dropped_cols.append(i)
        print("Dropping column ",i," Unique Value :",in_master_dataset[i].nunique)
    per = in_master_dataset[i].count()*100/tot_rcnt
    if per <= 90:
        in_master_dataset.drop(i,axis=1,inplace=True)
        dropped_cols.append(i)
        print("Dropping column ",i," % Values :",per)
```
