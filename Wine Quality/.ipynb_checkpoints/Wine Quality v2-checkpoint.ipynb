# Wine Quality Prediction

### Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

### Read File

```python
df_master = pd.read_csv("Datasets/winequality-red.csv", header=0)
df_master.head()
```

```python
df_master.isnull().sum()
df_master.describe()
```

```python
df_master.quality = df_master.quality.astype('category')
df_master.describe(include = 'all')
```

```python
df_master.columns
```

```python
for column in df_master:
    plt.figure()
    if df_master[column].dtype == "float64":
        f, (ax1, ax2, ax3) = plt.subplots(ncols = 3,figsize=(10,4))
        sn.kdeplot(df_master[column], shade=True, ax =ax1 ).set_title(column)
        sn.kdeplot(np.log(df_master[column]), shade=True, ax =ax2).set_title("log ("+column+")")
        sn.kdeplot(np.sqrt(df_master[column]), shade=True, ax =ax3).set_title("sqrt ("+column+")")
        
```

### Using the following transformed variables 

fixed acidity - log(fixed acidity)

volatile acidity - log(volatile acidity)

citric acid - log(citric acid)

residual sugar - log(residual sugar )

chlorides - log(chlorides )

free sulfur dioxide - ssqrt(free sulfur dioxide)

total sulfur dioxide - log(total sulfur dioxide)

density - log(density)

pH - pH

sulphates - log(sulphates)

alcohol - alcohol


```python
df_trans = df_master[["pH","alcohol","citric acid","quality"]].copy()
df_trans["fixed acidity"] = np.log(df_master["fixed acidity"])
df_trans["volatile acidity"] = np.log(df_master["volatile acidity"])
# df_trans["citric acid"] = np.log(df_master["citric acid"])
df_trans["residual sugar"] = np.log(df_master["residual sugar"])
df_trans["chlorides"] = np.log(df_master["chlorides"])
df_trans["free sulfur dioxide"] = np.sqrt(df_master["free sulfur dioxide"])
df_trans["density"] = np.log(df_master["density"])
df_trans["sulphates"] = np.log(df_master["sulphates"])
df_trans.head()
```

```python
for column in df_master:
    plt.figure()
    if df_master[column].dtype == "float64":
        sn.boxplot("quality", y=column, palette="Set3" , data = df_master )
    else:
        sn.barplot(x="quality",data = df_master)
```
