# Wine Quality Prediction

### Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

### Read File

```python
df_master = pd.read_csv("Datasets/winequality-red.csv", header=0)
df_master.head()
```

```python
df_master.isnull().sum()
df_master.describe()
```

```python
df_master.quality = df_master.quality.astype('category')
df_master.describe(include = 'all')
```

```python
for column in df_master:
    plt.figure()
    if df_master[column].dtype == "float64":
        sn.kdeplot(df_master[column], shade=True)
```

```python
for column in df_master:
    plt.figure()
    if df_master[column].dtype == "float64":
        sn.boxplot(x="quality", y=column, palette="Set3" , data = df_master )
```
